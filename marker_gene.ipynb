```python
import atlasapprox
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize the API
api = atlasapprox.API()
```

```python
organism = 'h_sapiens'
organ = 'lung'
cell_type = 'T'
```

```python
human_lung_T_markers = api.markers(organism=organism, organ=organ, cell_type=cell_type, number=10, measurement_type='gene_expression')
human_lung_T_markers
```

```python
human_lung_T_markers_exp = api.average(organism='h_sapiens', organ='lung', features=human_lung_T_markers, measurement_type='gene_expression')
human_lung_T_markers_exp
```

```python
human_lung_fraction = api.fraction_detected(organism='h_sapiens', organ='lung', features=human_lung_T_markers, measurement_type='gene_expression')
human_lung_fraction
```

```python
# Create the heatmap
plt.figure(figsize=(20, 6))
heatmap = sns.heatmap(
        human_lung_T_markers_exp, 
        cmap="Oranges", 
        annot=human_lung_fraction, 
        fmt=".2f", 
        linewidths=5,
        cbar_kws={"label": "Gene Expression Level"} 
        )

# Set labels
plt.title(f"Gene expression and fraction in {organism} {organ} by Cell Types", fontweight='bold')
plt.xlabel("Organs", fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.ylabel("Genes", fontsize=12, fontweight='bold')
cbar = heatmap.collections[0].colorbar
cbar.set_label("Gene Expression Level", fontweight="bold", fontsize=12)

# Show the plot
plt.show()
```
